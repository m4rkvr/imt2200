```python
import pandas as pd
```

```python
df_beneficiarios = pd.read_csv("../data/beneficiarios_sample.csv", encoding='latin1')
df_beneficiarios
```

```python
#revisar si tiene valores nulos
df_beneficiarios.info()
```

```python
#eliminar las filas con valores nulos
df_beneficiarios_sin_nulos= df_beneficiarios.dropna()
df_beneficiarios_sin_nulos.info()
```

```python
#verificar si existen valores duplicados
df_beneficiarios_sin_nulos.duplicated()
```

```python
#eliminar las filas duplicadas
df_beneficiarios_limpio = df_beneficiarios_sin_nulos.drop_duplicates()
df_beneficiarios_limpio.duplicated()
```

```python
df_beneficiarios_limpio
```

```python
#guardaremos el dataframe limpio en un archivo csv
df_beneficiarios_limpio.to_csv("../data/beneficiarios_sample_limpio.csv", encoding='latin1', index=False)
```
