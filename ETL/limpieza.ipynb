```python
import pandas as pd
```

## Primer dataset - Beneficiarios FONASA 2023

```python
original1 = False
try:
    with open("../data/originales/Beneficiarios Fonasa 2023.csv", encoding="latin1", errors="ignore") as f:
        df_beneficiarios = pd.read_csv(f, nrows=1000000)
        print("Se cargo la muestra original")
        original1 = True
        # De tener dataset original, indica numero de filas a cargar o quitar nrows para carga completa

except FileNotFoundError:
    print("El archivo original no se encuentra, abriremos muestra") # Muestra de 100000 filas, habilitado para subir a github
    df_beneficiarios = pd.read_csv("../data/sample/beneficiarios_sample.csv", encoding='latin1')
    print("Se cargo la muestra")
```

```python
df_beneficiarios
```

```python
#revisar si tiene valores nulos
df_beneficiarios.info()
```

```python
#eliminar las filas con valores nulos
df_beneficiarios_sin_nulos= df_beneficiarios.dropna()
df_beneficiarios_sin_nulos.info()
```

```python
#verificar si existen valores duplicados
df_beneficiarios_sin_nulos.duplicated()
```

```python
#eliminar las filas duplicadas
df_beneficiarios_limpio = df_beneficiarios_sin_nulos.drop_duplicates()
df_beneficiarios_limpio.duplicated()
```

```python
df_beneficiarios_limpio
```

```python
#guardaremos el dataframe limpio en un archivo csv
if original1 == True:
    df_beneficiarios_limpio.to_csv("../data/beneficiarios_original_limpio.csv", encoding='latin1', index=False)
elif original1 == False:
    df_beneficiarios_limpio.to_csv("../data/sample/beneficiarios_sample_limpio.csv", encoding='latin1', index=False)
```

## Tercer Dataframe - Presupuestos MLE 2023

```python
codigos_df = pd.read_excel("../data/originales/2 Planilla MLE 2023.xlsx", engine='openpyxl')
```

```python
# Eliminamos codigos repetidos
codigos_df = codigos_df[~codigos_df["total1"].isin([0])].copy()

# Formateamos codigos a GSSPPP
codigos_df["CODIGO_PRESTACIÓN"] = (
    codigos_df['grupo'].astype(str).str.zfill(1) +
    codigos_df['sub_grupo'].astype(str).str.zfill(2) +
    codigos_df['presta'].astype(str).str.zfill(3)
)
codigos_df = codigos_df.rename(columns={"glosa": "NOMBRE_PRESTACIÓN"})

nombre_codigos = codigos_df[["CODIGO_PRESTACIÓN", "NOMBRE_PRESTACIÓN"]]
```

```python
print(nombre_codigos['CODIGO_PRESTACIÓN'].duplicated().sum())
```

```python
codigos_df.sample(5)
```

```python
codigos_df.to_csv("../data/Presupuestos_MLE_2023_original_limpio.csv", encoding='latin1', index=False)
```

**Hacemos merge df_MLE_limpio y nombre_codigos; y añadimos columna "NOMRE_PRESTACION" al df_MLE_limpio**


```python
df_MLE_limpio
```

```python
df_MLE_limpio['CODIGO_PRESTACIÓN'] = df_MLE_limpio['CODIGO_PRESTACIÓN'].astype(str)
nombre_codigos["CODIGO_PRESTACIÓN"] = nombre_codigos["CODIGO_PRESTACIÓN"].astype(str)

df_MLE_limpio = df_MLE_limpio.merge(nombre_codigos, left_on="CODIGO_PRESTACIÓN", right_on="CODIGO_PRESTACIÓN", how='left')
```

```python
#guardaremos el dataframe limpio en un archivo csv
if original2 == True:
    df_MLE_limpio.to_csv("../data/MLE_2023_original_limpio.csv", encoding='latin1', index=False)
elif original2 == False:
    df_MLE_limpio.to_csv("../data/sample/MLE_2023_sample_limpio.csv", encoding='latin1', index=False)
```
